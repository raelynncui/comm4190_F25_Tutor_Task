So let’s tackle it in a way that makes sense, together.

In oil painting, “fat” means paint with more oil (which dries slower and stays flexible). “Lean” means paint with less oil (which dries faster and is stiffer). The rule says:

👉 Always paint “fat” layers on top of “lean” layers.  
Why? Because if a faster-drying (lean) layer is on top of a slower-drying (fat) one, the top layer can dry—and then crack—while the bottom is still shifting underneath. It’s like putting a crunchy shell on top of soft pudding—it’ll eventually break.
